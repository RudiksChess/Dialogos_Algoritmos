```python
def dataframe_nice(): 
  general = pd.DataFrame()
  for archivo in glob.glob("/content/DOCUMENTOS CONSOLIDADOS LIMPIOS/*.csv"): 
    nombre = str(os.path.basename(archivo))
    df = pd.read_csv(archivo,encoding='utf8',dtype=object)
    df["DB"] = nombre
    df["FECHA"]= pd.to_datetime(df["FECHA"],errors='coerce')
    importante = df[["FECHA","DB"]]
    importante = pd.DataFrame(importante)
    general=general.append(importante,ignore_index=True)
  
  return general
db= dataframe_nice()
```

```python
db = pd.get_dummies(db, columns=['DB'])
```

```python
db.head()
```

```python
cool= db.groupby([db.FECHA.dt.year,db.FECHA.dt.month]).sum()
```
